```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from tabulate import tabulate
```

# spectra without target

```python
datapoints = pd.read_csv("without-target-jan-25.csv",header=19)
counts = datapoints['Counts'].tolist()
channels = datapoints['Channel'].tolist()
```

```python
plt.figure(figsize=(20,5))
#plt.plot(channels, counts, '.')
plt.errorbar(channels, counts, yerr = 3*np.sqrt(counts), fmt='.b', ecolor='c', elinewidth = 1, capsize=2)
plt.annotate('Channel 414', xy=(414, 2334),  xycoords='data', xytext=(0, 100), textcoords='offset points', fontsize=16, \
            arrowprops=dict(color='r', shrink=0.05),ha='center', va='top')
plt.annotate('Channel 584', xy=(584, 974),  xycoords='data', xytext=(0, 100), textcoords='offset points', fontsize=16, \
            arrowprops=dict(color='r', shrink=0.05), ha='center', va='top')
plt.ylabel('Intensity [Count Number]', fontsize=16)
plt.xlabel('Energy [Channel Number]', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tick_params(bottom=True, top=True, left=True, right=True, direction="in")
#plt.grid()
#plt.legend()
plt.show()
```
